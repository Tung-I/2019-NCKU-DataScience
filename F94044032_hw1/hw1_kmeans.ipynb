```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
import math
```

```python
colors = ['r', 'g', 'b', 'orange']
df = pd.read_csv('cdata.csv')
point_x = df['x'].tolist()
point_y = df['y'].tolist()
data_len = len(point_x)
cluster_empty = [0]*data_len
df.head()
```

```python
def Distance(x1, y1, x2, y2):
    return math.sqrt(math.pow((x1-x2), 2)+math.pow((y1-y2), 2))

def SSE(center_x, center_y, cluster):
    sse = 0
    for i in range(data_len):
        idx_cen = cluster[i]
        d = Distance(point_x[i], point_y[i], center_x[idx_cen], center_y[idx_cen])
        sse+=pow(d, 2)
    return sse
        
# recalculate centers
def Re_center(cluster, k):
    num_cluster = len(cluster)
    center_x = []
    center_y = []
    for i in range(k):
        sum_x = 0
        sum_y = 0
        c = 0
        for j in range(data_len):
            if cluster[j]==i:
                sum_x+=point_x[j]
                sum_y+=point_y[j]
                c+=1
        if c==0:
            c=1
        center_x.append(float(sum_x)/c)
        center_y.append(float(sum_y)/c)
    return center_x, center_y

# initialiaze centers
def ini_center(k):
    center = []
    center_x = []
    center_y = []
    for i in range(k):
        while(True):
            idx = random.randint(0, data_len-1)
            judge = True
            for _center in center:
                d = Distance(_center[0], _center[1], point_x[idx], point_y[idx])
                if d<1:
                    judge = False
            if judge==True:
                center.append((point_x[idx], point_y[idx]))
                center_x.append(point_x[idx])
                center_y.append(point_y[idx])
                break
    return center_x, center_y

# iteration of Kmeans
def Kmeans_method(cen_x, cen_y, clu_cur):
    k = len(cen_x)
    clu_new = [0]*data_len
    # assign cluster
    for i in range(data_len):
        min_d = 99999
        for j in range(k):
            d = Distance(point_x[i], point_y[i], cen_x[j], cen_y[j])
            if d<min_d:
                min_d = d
                clu_new[i] = j
    # how many assignments have been changed
    num_change = 0
    for i in range(data_len):
        if clu_cur[i]!=clu_new[i]:
            num_change+=1
    # recalculate centers
    cen_x_new, cen_y_new = Re_center(clu_new, k)
    
    return cen_x_new, cen_y_new, clu_new, num_change

# divide into 4groups to plot
def Group4(cluster):
    group0_x = []
    group0_y = []
    group1_x = []
    group1_y = []
    group2_x = []
    group2_y = []
    group3_x = []
    group3_y = []
    for i in range(data_len):
        if cluster[i]==0:
            group0_x.append(point_x[i])
            group0_y.append(point_y[i])
        elif cluster[i]==1:
            group1_x.append(point_x[i])
            group1_y.append(point_y[i])
        elif cluster[i]==2:
            group2_x.append(point_x[i])
            group2_y.append(point_y[i])
        else:
            group3_x.append(point_x[i])
            group3_y.append(point_y[i])
    group_x = []
    group_y = []
    group_x.append(group0_x)
    group_x.append(group1_x)
    group_x.append(group2_x)
    group_x.append(group3_x)
    group_y.append(group0_y)
    group_y.append(group1_y)
    group_y.append(group2_y)
    group_y.append(group3_y)
    return group_x, group_y
        
def Kmeans(k):
    center_x, center_y = ini_center(k)
    cluster = cluster_empty.copy()
    while(True):
        center_x, center_y, cluster, num_change  = Kmeans_method(center_x, center_y, cluster)
        if num_change<5:
            break
    sse = SSE(center_x, center_y, cluster)
    return sse
```

Answer1

```python
k = 4
center_x, center_y = ini_center(k)
cluster = cluster_empty.copy()

while_cnt = 0
while(True):
    while_cnt+=1
    center_x, center_y, cluster, num_change  = Kmeans_method(center_x, center_y, cluster)
    # grouping to plot        
    group_x, group_y = Group4(cluster)
    # plot
    plt.figure()
    for i in range(k):
        plt.scatter(group_x[i], group_y[i], c = colors[i], label = str(i))
    plt.scatter(center_x, center_y, c = 'black', marker = 'x', label = 'center')
    plt.ylabel('y')
    plt.xlabel('x')
    plt.legend(labels = ['A', 'B', 'C', 'D', 'Central'])
    plt.title("Round"+str(while_cnt))
    # if converge
    if num_change<5:
        break



```

```python
sse_list = []
for k in range(2, 11):
    sse = Kmeans(k)
    sse_list.append(sse)
plt.figure()
plt.ylabel('Sum of square error')
plt.xlabel('Number of k')
plt.plot(range(2, 11), sse_list)
    
```
